```python
def funksvd(matrix,latent,lr,iteration):
    np.random.seed(52)
    users = matrix.shape[0]
    movies = matrix.shape[1]
    squared_error = 0
    user = np.random.rand(users,latent)
    movie = np.random.rand(latent,movies)
    number_of_ratings=100000
    print("Iterations | Mean Squared Error ")
    for i4 in range(iteration):
        squared_error = 0
        for i in range(users):
            for j in range(movies):
                if (matrix[i, j] !=0.0):
                    difference = matrix[i, j] - np.dot(user[i, :], movie[:, j])
                    squared_error=squared_error+difference**2
                    for k in range(latent):
                        user[i, k] += lr * (2*difference*movie[k, j])
                        movie[k, j] += lr * (2*difference*user[i, k])
        print("%d \t\t %f" % (i4+1,squared_error/ number_of_ratings))
    
    return movie,user
```
